```python
def group_data(df, column_name):
    
    # Group the data by the column_name
    grouped_data = df.groupby(by=column_name)
    data = []

    # Iterate over grouped data and store the block index and number of lots the block has in block_data.
    for row in grouped_data:
        data.append((row[0], len(row[1])))
        
    return data
```

#### Column 4: LotSize

Lot size is the dirtiest column present in the dataset:
    
    1) It contains the area of the lot but the formatting is not uniform. 
    2) Area can be in acres, SQ. FT. and can be in an unknown unit format represented by multiplication "X" sign between    numbers e.g. 18X55.

Let's do some further analysis on this column.

Find what are the different formatting other than acres, SQ. FT. and unknown.

    1) For the unknown unit format I have to check for "X" in the string.
    2) For the Square Feet unit format check for character "S" and "F".
    3) For Acres unit format check for "ACRE" keyword in the string.
    4) Datapoints which will not satisfy above conditions must be in different formatting. 

```python

for row in df.iterrows():
    idx, data = row
    if data.LotSize.find("X")==-1 and data.LotSize.find("ACRES")==-1 \
    and (data.LotSize.find("S")==-1 or data.LotSize.find("F")==-1):
                print(data.LotSize)

    
```

It seems if we take AC keyword instead of ACRES we can cover all misspelled ACRES.

To make the LotSize column usable for further processing I need to make it a numeric column by converting all different format datapoints to float with Sq Ft as their units.

```python
# A counter to count number of unhandled values
count = 0

# Declaring all the lists for further processing.
acres_area = []
sqft_area = []
unk_area = []
areas = []

# Iterate over data frame
for row in df.iterrows():
    
    idx, data = row
    area = data.LotSize
    
    # check for unknown format data
    if area.find("X")!=-1:
        
        # Convert unknown data into float values
        try:
            area_vals = area.split(maxsplit=-1)[0].split("X")
            area_vals = list(map(lambda x: x.strip().replace("-", "."),area_vals[:2]))
            area = float(area_vals[0])*float(area_vals[1])
            unk_area.append(area)
        except:
            print(area)
            count += 1
            area = -1
            
    # Check for AC keyword if found then find the numeric part and multiply it by 43560 to convert it in sqft.
    elif area.find("AC")!=-1:
        pattern = re.compile("[^\W\d]")
        area = area.replace("O","0")
        match = pattern.search(area)
        splited = [area[:match.start()], area[match.start():]]
        area = float(splited[0].strip().replace("-", "."))
        acres_area.append(area)  
        
    # Check for S and F characters for sqft.
    elif area.find("S")!=-1 and area.find("F")!=-1:
        area = float(area.split("S")[0].strip().replace(",","").replace("\\",""))
        sqft_area.append(area)

    else:
        print(area)
        area = -1
        count += 1      
       
    areas.append(area)
print("Total rejected values: {}".format(count))
    
```

The above printed values are rejected as the don't have proper format. There are total  47 rejected rows

Now, we have to check whether unknown unit data is in acres or in SQ. FT.

```python
print("Unknown Values: Min- {}, Max- {}, Mean- {}".format(np.min(unk_area),np.max(unk_area),np.mean(unk_area)))
print("Sq ft Values  : Min- {}, Max- {}, Mean- {}".format(np.min(sqft_area),np.max(sqft_area),np.mean(sqft_area)))
print("Acres Values  : Min- {}, Max- {}, Mean- {}".format(np.min(acres_area),np.max(acres_area),np.mean(acres_area)))
```

It can be inferred from the above mean values that unknown values parameter are somewhat close to Sq ft values hence we will assume that unknown values are Sq ft values.

Also, we will convert the acre values to Sq. Ft. by multiplying them by 43560

```python
count = 0
for row in df.iterrows():
    
    idx, data = row
    area = data.LotSize
    if area.find("AC")!=-1:
        pattern = re.compile("[^\W\d]")
        area = area.replace("O","0")
        match = pattern.search(area)
        splited = [area[:match.start()], area[match.start():]]
        area = float(splited[0].strip().replace("-", "."))*43560
        areas[count] = area
    count += 1
```

We will add this new column to our dataset under 'Areas' name.

```python
df["Areas"] = areas
```

Now we will drop the columns which don't have proper LotSize format

```python
df = df[df.Areas != -1]
```

As there are more than 3000 block in the dataset, so to maintain the readability I will just take top three blocks with the  most number of lots per ward.

```python
grouped = df1.groupby("Ward")
wards = sorted(df["Ward"].unique())
blocks_for_plotting = []
for i in wards:
    grouped_by_block = group_data(df1[df1.Ward==i], "Block")
    sorted_block_data = sorted(grouped_by_block, key= lambda x: x[1])[-3:]
    print(sorted_block_data)
    blocks_for_plotting.append(list(zip(*sorted_block_data))[0])
```
